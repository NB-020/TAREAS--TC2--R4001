#### PUNTO #1

**(a)** Para poder sintetizar el cuadripolo pedido, lo primero a realizar será encontrar la forma de la transferencia de tensiones presentada en función de los parámetros correspondientes, que podrán ser Z o Y. Para ello, planteamos lo siguiente:

\begin{equation}
\frac{V_2}{V_g}|_{I_2 = 0} = \left[\frac{V_2}{V_1} \cdot \frac{V_1}{V_g} \right]|_{I_2 = 0} = \frac{Z_{21}}{Z_{11}} \cdot \frac{Z_{11}}{Z_{11} + Z_g} = \frac{Z_{21}}{Z_{11} + Z_g}
\end{equation}

De esta forma, encontramos que la transferencia de tensiones solicitada se encuentra en función de los parámetros Z del cuadripolo, y de la impedancia del generador. Luego, como la síntesis es sin pérdidas, entonces consideraremos unitaria a la resistencia del generador ($R_g = 1\Omega$), de forma que:

\begin{equation}
\frac{V_2}{V_g}|_{I_2 = 0} = \frac{Z_{21}}{1 + Z_{11}}
\end{equation}

Esto nos determina que la red será sintetizada desde la entrada, hacia la salida, ya que buscaremos sintetizar $Z_{11}$. Lo siguiente será identificar qué parte de la transferencia dada se corresponde con $Z_{21}$, y qué otra parte con $Z_{11}$. Para ello, utilizando la expresión brindada de $T(s)$, debido a que el numerador es un polinomio impar, sacamos de factor común del denominador solo la parte par, garantizando imparidad de la función transferencia, quedando de esta manera:

\begin{equation}
T(s) = \frac{k \cdot s \cdot (s^2 + \frac{1}{16})}{s^3 + 2s^2 + 2s + 1} = \frac{ \frac{k \cdot s \cdot (s^2 + \frac{1}{16})}{2s^2 + 1} }{ 1 + \frac{s^3 + 2s}{2s^2 + 1} } = \frac{Z_{21}}{1 + Z_{11}}
\end{equation}

\begin{equation}
\boxed{Z_{21} = \frac{k \cdot s \cdot (s^2 + \frac{1}{16})}{2s^2 + 1}} \quad ; \quad \boxed{Z_{11} = \frac{s^3 + 2s}{2s^2 + 1}}
\end{equation}

Ahora, se procede a realizar el Método Gráfico para determinar la topología del cuadripolo. Debido a que no tenemos condiciones de medición en la entrada, el primer componente, visto desde la entrada, podrá estar en serie o en derivación, pero el último deberá estar necesariamente en derivación, ya que la condición de medición en la salida es $I_2 = 0$. De esta forma, se procede al método gráfico y cálculo de los componentes:

![alt text](met_graf_ej1.png)

Primero invertimos $Z_0 = Z_{11}$ para encontrar $Y_0$, y luego realizamos una remoción parcial en DC, hallando su residuo correspondiente y obteniendo un inductor en derivación:

\begin{equation}
Z_0(s) = Z_{11}(s) = \frac{s^3 + 2s}{2s^2 + 1} \quad \rightarrow \quad Y_0(s) = \frac{2s^2 + 1}{s^3 + 2s} = \frac{2 \cdot (s^2 + \frac{1}{2})}{s \cdot (s^2 + 2)}
\end{equation}

\begin{equation}
Y_2(s) =  Y_0(s) - Y_1(s) \quad ; \quad Y_1(s) = \frac{k_1}{s}
\end{equation}

\begin{equation}
k_1 = \lim_{s^2 \to -\frac{1}{16}} Y_0(s) \cdot s = \frac{14}{31} \quad ; \quad \boxed{Y_1(s) = \frac{ \frac{14}{31} }{s}}
\end{equation}

\begin{equation}
Y_2(s) = Y_0(s) - Y_1(s) = \frac{ \frac{48}{31} s^2 + \frac{3}{31} }{ s^3 + 2s } = \frac{ \frac{48}{31} \cdot (s^2 + \frac{1}{16}) }{ s \cdot (s^2 + 2) }
\end{equation}

El siguiente paso, será invertir $Y_2(s)$ para obtener $Z_2(s)$ y realizar una remoción total en $\frac{1}{4}$, obteniendo un tanque LC en serie:

\begin{equation}
Z_2(s) = \frac{ s^3 + 2s }{ \frac{48}{31} s^2 + \frac{3}{31} }
\end{equation}

\begin{equation}
Z_4(s) =  Z_2(s) - Z_3(s) \quad ; \quad Z_3(s) = \frac{2k_3 \cdot s}{s^2 + \frac{1}{16}}
\end{equation}

\begin{equation}
2k_3 = \lim_{s^2 \to -\frac{1}{16}} Z_2(s) \cdot \frac{s^2 + \frac{1}{16}}{s} = \frac{961}{768} \quad ; \quad \boxed{Z_3(s) = \frac{\frac{961}{768} \cdot s}{s^2 + \frac{1}{16}}}
\end{equation}

\begin{equation}
\boxed{Z_4(s) = Z_2(s) - Z_3(s) = \frac{31}{48} \cdot s}
\end{equation}

\end{equation}

\begin{equation}
\boxed{\frac{V_2}{I_g}|_{I_2 = 0} = \frac{ s^3 + \frac{1}{16} \cdot s }{ s^3 + 2 s^2 + 2 s + 1 }}
\end{equation}

Al igual que en el ejercicio (1), según la verificación, el valor de "k" debería ser 1. Sin embargo, como se pide que la transferencia presente una magnitud de -6 dB en el centro de la banda de paso, entonces forzaremos el valor de "k", encontrando el módulo de la respuesta en frecuencia del cuadripolo, y determinando su valor en magnitud para el centro de la banda de paso:

\begin{equation}
T(s) = \frac{k \cdot s \cdot (s^2 + \frac{1}{16})}{s^3 + 2s^2 + 2s + 1}
\end{equation}

\begin{equation}
T(\omega) = T(s)|_{s = j\omega} = \frac{k \cdot j \cdot (\frac{\omega}{16} - \omega^3)}{(1 - 2\omega^2) + j \cdot (2\omega - \omega^3)}
\end{equation}

\begin{equation}
|T(\omega)| = \frac{k \cdot |\frac{\omega}{16} - \omega^3|}{ \sqrt{(1 - 2\omega^2)^2 + (2\omega - \omega^3)^2} }
\end{equation}

Debido a que la transimpedancia presentada puede entenderse o imaginarse como una transferencia PASA ALTOS, entonces el centro de la banda de paso estará dado cuando $\omega \to \infty$, es decir, para alta frecuencia. En dicha banda, la transferencia tiende a 0 dB. De esta forma, el valor de "k" será:

\begin{equation}
\boxed{k = \lim_{\omega \to \infty} |T(\omega)| = -6 dB = 0.5012}
\end{equation}
