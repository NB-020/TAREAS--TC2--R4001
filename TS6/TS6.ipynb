#### PUNTO #1

$\rightarrow$ En primer lugar, analizaremos la red Tee del primer circuito. Los parámetos elegidos para ello fueron los $Z$ de impedancia, cuya matriz y ecuaciones constitutivas son las siguientes:

$\hspace{3cm}$![alt text](red_t_ts6.jpg)

\begin{equation}
Z = \begin{pmatrix}
    Z_{11} & Z_{12} \\
    Z_{21} & Z_{22}
    \end{pmatrix}
\end{equation}

\begin{equation}
V_1 = I_1 \cdot Z_{11} + I_2 \cdot Z_{12} \quad ; \quad V_2 = I_1 \cdot Z_{21} + I_2 \cdot Z_{22}
\end{equation}

Siendo $Z_{11}$ la **impedancia de excitación de entrada con la salida en vacío**; $Z_{12}$ la **impedancia de transferencia inversa con la entrada en vacío**; $Z_{21}$ la **impedancia de transferencia directa con la salida en vacío**; $Z_{22}$ la **impedancia de excitación de salida con la entrada en vacío**.

De esta manera, podemos determinar cada uno de los parámetros:

\begin{equation}
Z_{11} = \frac{V_1}{I_1}|_{I_2 = 0} = Z_a + Z_b
\end{equation}

\begin{equation}
Z_{12} = \frac{V_1}{I_2}|_{I_1 = 0} = Z_b
\end{equation}

\begin{equation}
Z_{21} = \frac{V_2}{I_1}|_{I_2 = 0} = Z_b
\end{equation}

\begin{equation}
Z_{22} = \frac{V_2}{I_2}|_{I_1 = 0} = Z_b + Z_c
\end{equation}

\begin{equation}
Z = \begin{pmatrix}
    Z_a + Z_b & Z_b \\
    Z_b & Z_b + Z_c
    \end{pmatrix}
\end{equation}


$\rightarrow$ La siguiente red será la red $\pi$ o Pi. En este caso, se eligío trabajar con los parámetros $Y$ de admitancia, cuya matriz y ecuaciones constitutivas se detallan a continuación:

$\hspace{3cm}$![alt text](red_pi_ts6.jpg)

\begin{equation}
Y = \begin{pmatrix}
    Y_{11} & Y_{12} \\
    Y_{21} & Y_{22}
    \end{pmatrix}
\end{equation}

\begin{equation}
I_1 = V_1 \cdot Y_{11} + V_2 \cdot Y_{12} \quad ; \quad I_2 = V_1 \cdot Y_{21} + V_2 \cdot Y_{22}
\end{equation}

Siendo $Y_{11}$ la **admitancia de entrada con la salida en cortocircuito**; $Y_{12}$ la **admitancia de transferencia inversa con la entrada en cortocircuito**; $Y_{21}$ la **admitancia de transferencia directa con la salida en cortocircuito**; $Y_{22}$ la **admitancia de salida con la entrada en cortocircuito**.

Así, los parámetros pueden determinarse de la siguiente forma:

\begin{equation}
Y_{11} = \frac{I_1}{V_1}|_{V_2 = 0} = Y_a + Y_c
\end{equation}

\begin{equation}
Y_{12} = \frac{I_1}{V_2}|_{V_1 = 0} = -Y_c
\end{equation}

\begin{equation}
Y_{21} = \frac{I_2}{V_1}|_{V_2 = 0} = -Y_c
\end{equation}

\begin{equation}
Y_{22} = \frac{I_2}{V_2}|_{V_1 = 0} = Y_b + Y_c
\end{equation}

\begin{equation}
Y = \begin{pmatrix}
    Y_a + Y_c & -Y_c \\
    -Y_c & Y_b + Y_c
    \end{pmatrix}
\end{equation}


$\rightarrow$ Como siguiente circuito se encuentra la red Lattice, para la cual se analizaron los parámetros $Z$:

$\hspace{3cm}$![alt text](red_lattice_ts6.jpg)

\begin{equation}
Z = \begin{pmatrix}
    Z_{11} & Z_{12} \\
    Z_{21} & Z_{22}
    \end{pmatrix}
\end{equation}

\begin{equation}
V_1 = I_1 \cdot Z_{11} + I_2 \cdot Z_{12} \quad ; \quad V_2 = I_1 \cdot Z_{21} + I_2 \cdot Z_{22}
\end{equation}

Al ser una red simétrica, se puede determinar que $Z_{11} = Z_{22}$ y $Z_{12} = Z_{21}$. De esta forma, el análisis será el siguiente:

Para $Z_{11}$, al alimentar el circuito con una corriente $I_1$ y dejar los puertos de salida abiertos ($I_2 = 0$), la corriente de entrada se divide en dos caminos hasta masa: uno a través de $Z_a$ en la parte superior y por $Z_b$ en diagonal de derecha a izquierda; y otro a través $Z_b$ en diagonal de izquierda a derecha y luego por $Z_a$ en la parte inferior. Al haber la misma impedancia en ambos caminos, la corriente circulando en cada uno de ellos será $\frac{I_1}{2}$, y así se puede determinar:

\begin{equation}
Z_{11} = \frac{V_1}{I_1}|_{I_2 = 0} = Z_{22} = \frac{V_2}{I_2}|_{I_1 = 0} = \frac{Z_a + Z_b}{2}
\end{equation}

Luego, al igual que en el caso anterior, si alimentamos la red con una corriente $I_2$ por los puertos de salida, dejando los de entrada abiertos, dicha corriente se divide en dos partes iguales por misma impedancia en ambos caminos de circulación hasta masa. Por ende, resulta:

\begin{equation}
Z_{12} = \frac{V_1}{I_2}|_{I_1 = 0} = Z_{21} = \frac{V_2}{I_1}|_{I_2 = 0} = \frac{Z_a + Z_b}{2}
\end{equation}

\begin{equation}
Z = \begin{pmatrix}
    \frac{Z_a + Z_b}{2} & \frac{Z_a + Z_b}{2} \\
    \frac{Z_a + Z_b}{2} & \frac{Z_a + Z_b}{2}
    \end{pmatrix}
\end{equation}


$\rightarrow$ El siguiente circuito se encuentra conformado únicamente por un OTA (Operational Transconductance Amplifier), para el cual se eligío trabajar con parámetros T, cuya matriz y ecuaciónes constitutivas serán:

$\hspace{3cm}$![alt text](red_ota_ts6.jpg)

\begin{equation}
T = \begin{pmatrix}
    A & B \\
    C & D
    \end{pmatrix}
\end{equation}

\begin{equation}
V_1 = V_2 \cdot A + (-I_2) \cdot B \quad ; \quad I_1 = V_2 \cdot C + (-I_2) \cdot D
\end{equation}

Siendo $A$ la **inversa de la transferencia de tensiones en vacío**; $B$ la **inversa de la transferencia directa corriente-tensión en cortocircuito**; $C$ la **inversa de la transferencia directa tensión-corriente en vacío**; $D$ la **inversa de la transferencia directa de corrientes en cortocircuito**.

Siguiendo, los parámetros de la matriz de determinan de la siguiente forma:

\begin{equation}
A = \frac{V_1}{V_2}|_{I_2 = 0} = \frac{1}{\frac{V_2}{V_1}}|_{I_2 = 0} = 0
\end{equation}

\begin{equation}
B = \frac{V_1}{(-I_2)}|_{V_2 = 0} = \frac{1}{\frac{(-I_2)}{V_1}}|_{V_2 = 0} = \frac{1}{g_m}
\end{equation}

\begin{equation}
C = \frac{I_1}{V_2}|_{I_2 = 0} = \frac{1}{\frac{V_2}{I_1}}|_{I_2 = 0} = 0
\end{equation}

\begin{equation}
D = \frac{I_1}{(-I_2)}|_{V_2 = 0} = \frac{1}{\frac{(-I_2)}{I_1}}|_{V_2 = 0} = 0
\end{equation}

\begin{equation}
T = \begin{pmatrix}
    0 & \frac{1}{g_m} \\
    0 & 0
    \end{pmatrix}
\end{equation}


$\rightarrow$ El próximo circuito esta integrado por un Amplificador Operacional (OPAMP) en su modalidad de inversor. Los parámetros elegidos para su análisis fueron los $Z$. Se detallan a continuación las ecuaciónes descriptivas del circuito:

$\hspace{3cm}$![alt text](red_opamp_ts6.jpg)

\begin{equation}
I_1 = \frac{V_1}{R_1} \quad ; \quad I_1 = -\frac{V_2}{R_2}

\begin{equation}
Z = \begin{pmatrix}

\begin{equation}

\end{equation}

A continuación, se determinan los parámetros $Z$:

\begin{equation}
Z_{11} = \frac{V_1}{I_1}|_{I_2 = 0} = R_1

\begin{equation}
Z_{12} = \frac{V_1}{I_2}|_{I_1 = 0} = 0

Z_{21} = \frac{V_2}{I_1}|_{I_2 = 0} = -R_2

$\rightarrow$ Como última red, se encuentra un transformador ideal, del cual podemos determinar las siguientes ecuaciones descriptivas. Se decidió determinar los parámetros T de este circuito.

$\hspace{3cm}$![alt text](red_trafo_ts6.jpg)

\begin{equation}
V_1 = N \cdot V_2 \quad ; \quad I_1 = \frac{(-I_2)}{N} \quad ; \quad \frac{V_1}{V_2} = \frac{(-I_2)}{I_1} = N = a
\end{equation}

\begin{equation}
T = \begin{pmatrix}
    A & B \\
    C & D
    \end{pmatrix}
\end{equation}

\begin{equation}
V_1 = V_2 \cdot A + (-I_2) \cdot B \quad ; \quad I_1 = V_2 \cdot C + (-I_2) \cdot D
\end{equation}

Luego, los parámetros T se pueden determinar como sigue:

\begin{equation}
A = \frac{V_1}{V_2}|_{I_2 = 0} = \frac{1}{\frac{V_2}{V_1}}|_{I_2 = 0} = N = a
\end{equation}

\begin{equation}
B = \frac{V_1}{(-I_2)}|_{V_2 = 0} = \frac{1}{\frac{(-I_2)}{V_1}}|_{V_2 = 0} = 0
\end{equation}

\begin{equation}
C = \frac{I_1}{V_2}|_{I_2 = 0} = \frac{1}{\frac{V_2}{I_1}}|_{I_2 = 0} = 0
\end{equation}

\begin{equation}
D = \frac{I_1}{(-I_2)}|_{V_2 = 0} = \frac{1}{\frac{(-I_2)}{I_1}}|_{V_2 = 0} = \frac{1}{N} = \frac{1}{a}
\end{equation}

\begin{equation}
T = \begin{pmatrix}
    N & 0 \\
    0 & \frac{1}{N}
    \end{pmatrix}   =   \begin{pmatrix}
                        a & 0 \\
                        0 & \frac{1}{a}
                        \end{pmatrix}
\end{equation}

#### PUNTO #2:

Para este punto será conveniente saber cómo pasar de un tipo de parámetro a otro, aplicando las transformaciones. Esto es posible si elegimos los parámetros a los que queremos llegar y las ecuaciones constitutivas de los parámetros de los cuales partimos, y por medio de operaciones y procedimientos matemáticos y algebráicos, se llega a las expresiones de transformación. A continuación se detallan las transformaciones de parámetros $Z$ en parámetros $Y$ y $T$, y viceversa:

$\rightarrow$ **PARÁMETROS Z A PARÁMETROS Y**

\begin{equation}
Z = \begin{pmatrix}
    Z_{11} & Z_{12} \\
    Z_{21} & Z_{22}
    \end{pmatrix}   \rightarrow     Y = Z^{-1} = \frac{Adj(Z)}{\Delta Z}
\end{equation}

$\rightarrow$ **PARÁMETROS Z A PARÁMETROS T**

\begin{equation}
Z = \begin{pmatrix}
    Z_{11} & Z_{12} \\
    Z_{21} & Z_{22}
    \end{pmatrix}   \rightarrow     T = \begin{pmatrix}
                                        A & B \\
                                        C & D
                                        \end{pmatrix}   =  \frac{1}{Z_{21}} \begin{pmatrix}
                                                                            Z_{11} & \Delta Z \\
                                                                            1 & Z_{22}
                                                                            \end{pmatrix}
\end{equation}

$\rightarrow$ **PARÁMETROS Y A PARÁMETROS Z**

\begin{equation}
Y = \begin{pmatrix}
    Y_{11} & Y_{12} \\
    Y_{21} & Y_{22}
    \end{pmatrix}   \rightarrow     Z = Y^{-1} = \frac{Adj(Y)}{\Delta Y}
\end{equation}

$\rightarrow$ **PARÁMETROS Y A PARÁMETROS T**

\begin{equation}
Y = \begin{pmatrix}
    Y_{11} & Y_{12} \\
    Y_{21} & Y_{22}
    \end{pmatrix}   \rightarrow     T = \begin{pmatrix}
                                        A & B \\
                                        C & D
                                        \end{pmatrix}   =  \frac{-1}{Y_{21}} \begin{pmatrix}
                                                                            Y_{22} & 1 \\
                                                                            \Delta Y & Y_{11}
                                                                            \end{pmatrix}
\end{equation}

$\rightarrow$ **PARÁMETROS T A PARÁMETROS Z**

\begin{equation}
T = \begin{pmatrix}
    A & B \\
    C & D
    \end{pmatrix}   \rightarrow     Z = \begin{pmatrix}
                                        Z_{11} & Z_{12} \\
                                        Z_{21} & Z_{22}
                                        \end{pmatrix}   =  \frac{1}{C} \begin{pmatrix}
                                                                        A & \Delta T \\
                                                                        1 & D
                                                                        \end{pmatrix}
\end{equation}

$\rightarrow$ **PARÁMETROS T A PARÁMETROS Y**

\begin{equation}
T = \begin{pmatrix}
    A & B \\
    C & D
    \end{pmatrix}   \rightarrow     Y = \begin{pmatrix}
                                        Y_{11} & Y_{12} \\
                                        Y_{21} & Y_{22}
                                        \end{pmatrix}   =  \frac{1}{B} \begin{pmatrix}
                                                                        D & -\Delta T \\
                                                                        -1 & A
                                                                        \end{pmatrix}
\end{equation}

Sabiendo esto, podemos hacer las transformaciónes de los circuitos analizados anteriormente:


$\rightarrow$ Red Tee:

$\hspace{3cm}$![alt text](red_t_ts6.jpg)

\begin{equation}
Z = \begin{pmatrix}
    Z_{11} & Z_{12} \\
    Z_{21} & Z_{22}
    \end{pmatrix}   =   \begin{pmatrix}
                        Z_a + Z_b & Z_b \\
                        Z_b & Z_b + Z_c
                        \end{pmatrix}
\end{equation}

\begin{equation}
\Delta Z = Z_a \cdot Z_b + Z_a \cdot Z_c + Z_b \cdot Z_c \quad ; \quad Adj(Z) = \begin{pmatrix}
                                                                                Z_b + Z_c & -Z_b \\
                                                                                -Z_b & Z_a + Z_b
                                                                                \end{pmatrix}
\end{equation}

\begin{equation}
Y = \begin{pmatrix}
    Y_{11} & Y_{12} \\
    Y_{21} & Y_{22}
    \end{pmatrix}    = Z^{-1} = \frac{Adj(Z)}{\Delta Z} = \frac{1}{Z_a \cdot Z_b + Z_a \cdot Z_c + Z_b \cdot Z_c} \cdot \begin{pmatrix}
                                                                                                                        Z_b + Z_c & -Z_b \\
                                                                                                                        -Z_b & Z_a + Z_b
                                                                                                                        \end{pmatrix}
\end{equation}

\begin{equation}
T = \begin{pmatrix}
    A & B \\
    C & D
    \end{pmatrix} = \frac{1}{Z_b} \cdot \begin{pmatrix}
                                        Z_a + Z_b & Z_a \cdot Z_b + Z_a \cdot Z_c + Z_b \cdot Z_c \\
                                        1 & Z_b + Z_c
                                        \end{pmatrix}
\end{equation}


$\rightarrow$ Red $\pi$:

$\hspace{3cm}$![alt text](red_pi_ts6.jpg)

\begin{equation}
Y = \begin{pmatrix}
    Y_{11} & Y_{12} \\
    Y_{21} & Y_{22}
    \end{pmatrix} = \begin{pmatrix}
                    Y_a + Y_c & -Y_c \\
                    -Y_c & Y_b + Y_c
                    \end{pmatrix}
\end{equation}

\begin{equation}
\Delta Y = Y_a \cdot Y_b + Y_a \cdot Y_c + Y_b \cdot Y_c \quad ; \quad Adj(Y) = \begin{pmatrix}
                                                                                Y_b + Y_c & Y_c \\
                                                                                Y_c & Y_a + Y_c
                                                                                \end{pmatrix}
\end{equation}

\begin{equation}
Z = \begin{pmatrix}
    Z_{11} & Z_{12} \\
    Z_{21} & Z_{22}
    \end{pmatrix}    = Y^{-1} = \frac{Adj(Y)}{\Delta Y} = \frac{1}{Y_a \cdot Y_b + Y_a \cdot Y_c + Y_b \cdot Y_c} \cdot \begin{pmatrix}
                                                                                                                        Y_b + Y_c & Y_c \\
                                                                                                                        Y_c & Y_a + Y_c
                                                                                                                        \end{pmatrix}
\end{equation}

\begin{equation}
T = \begin{pmatrix}
    A & B \\
    C & D
    \end{pmatrix} = \frac{1}{Y_c} \cdot \begin{pmatrix}
                                        Y_b + Z_c & 1 \\
                                        Y_a \cdot Y_b + Y_a \cdot Y_c + Y_b \cdot Y_c & Y_a + Y_c
                                        \end{pmatrix}
\end{equation}


$\rightarrow$ Red Lattice:

$\hspace{3cm}$![alt text](red_lattice_ts6.jpg)

\begin{equation}

$\rightarrow$ Red OTA:

$\hspace{3cm}$![alt text](red_ota_ts6.jpg)

\begin{equation}

    \end{pmatrix} = \begin{pmatrix}

    \end{pmatrix} = \begin{pmatrix}

                    \end{pmatrix}
\end{equation}

\begin{equation}
\Delta Z = 0 \quad ; \quad Adj(Z) = \begin{pmatrix}
                                    0 & 0 \\

                                    \end{pmatrix}
\end{equation}

Debido a que el determinante de la matriz de parámetros $Z$ es cero, los parámetros $Y$ no se encuentran definidos para esta red. Sin embargo, los parámetros T serán:

\begin{equation}
T = \begin{pmatrix}
    A & B \\
    C & D
    \end{pmatrix} = \frac{1}{(-R_2)} \cdot \begin{pmatrix}

    \end{pmatrix} = \begin{pmatrix}

Según las expresiones de transformación obtenidas más arriba, debido a que los parámetros $C$ y $B$ son nulos en la matriz de parámetros $T$, los parámetros $Z$ y $Y$, de impedancia y admitancia respectivamente, no se encuentran definidos para un transformador ideal como de plantea en la figura.
