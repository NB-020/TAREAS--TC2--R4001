#### PUNTO #1

A través del siguiente análisis del circuito planteado en el enunciado, podemos determinar las ecuaciones que nos ayudarán a determinar la función transferencia de la red en cuestión:

![alt text](ackerberg_mossberg_analisis.png)

\begin{equation}
\frac{V_i}{R_1} + V_x(sC + \frac{1}{R_2}) + \frac{V_o}{R_3} = 0
\end{equation}

\begin{equation}
\frac{V_x}{R_3} + V_y sC = 0
\end{equation}

\begin{equation}
\frac{V_y}{R_4} + \frac{V_o}{R_4} = 0
\end{equation}

Despejando convenientemente, podemos obtener las siguientes expresiones:

\begin{equation}
V_y = -V_o
\end{equation}

\begin{equation}
V_x = V_o sCR_3
\end{equation}

Por último, reemplazamos en la primera ecuación $V_x$ por su valor en función de $V_o$ y volvemos a despejar para obtener la siguiente transferencia:

\begin{equation}
\frac{V_i}{R_1} = -V_o sCR_3(\frac{sCR_2+1}{R_2}) - \frac{V_o}{R_3}
\end{equation}

\begin{equation}
\frac{V_i}{R_1} = -V_o(\frac{s^2 C^2 R_2R_3^2 + sCR_3^2 + R_2}{R_2R_3})
\end{equation}

\begin{equation}
H(s) = \frac{V_o}{V_i} = (-1) \frac{\frac{1}{C^2R_1R_3}}{s^2 + s \frac{1}{CR_2} + \frac{1}{C^2R_3^2}}
\end{equation}

Por inspección de esta última expresión, podemos determinar que al tomar la salida en la tensión $V_o$ obtenemos un **Filtro Pasa Bajos**, el cual se corresponde con:

\begin{equation}
H(s) = \frac{V_o}{V_i} = k \frac{\omega_0^2}{s^2 + s \frac{\omega_0}{Q} + \omega_0^2}
\end{equation}

#### PUNTO #3:

Para obtener un **Filtro Pasa Banda** con la red vista, solo basta con observar el circuito nuevamente. Si analizamos el nodo $V_x$, podemos determinar que el capacitor que se encuentra en paralelo con la resistencia $R_2$, a frecuencia $\omega = 0$ se comporta como un circuito abierto, y el operacional $U_1$ actúa como un inversor. Luego, para frecuencias tendiendo a infinito, el capacitor es un cortocircuito, provocando que $V_x$ pase a tener los 0V virtuales del operacional.

Dicho esto, procedemos a tomar la salida en el nodo de tensión $V_x$, y con las ecuaciones vistas en el **Punto #1** obtenemos:

\begin{equation}
V_y = -V_o
\end{equation}

\begin{equation}
V_0 = \frac{V_x}{sCR_3}
\end{equation}

Luego, llegamos a las siguientes expresiones:

\begin{equation}
\frac{V_i}{R_1} = -V_x(\frac{s^2 C^2 R_2R_3^2 + sCR_3^2 + R_2}{sCR_2R_3^2})
\end{equation}

\begin{equation}
H_x(s) = \frac{V_x}{V_i} = (-1) \frac{s\frac{1}{CR_1}}{s^2 + s \frac{1}{CR_2} + \frac{1}{C^2R_3^2}}
\end{equation}

Esta última se corresponde con la forma de un **Filtro Pasa Banda**, que es el buscado:

\begin{equation}
H_x(s) = \frac{V_o}{V_i} = k \frac{s \frac{\omega_0}{Q}}{s^2 + s \frac{\omega_0}{Q} + \omega_0^2}
\end{equation}

Relacionando las últimas dos expresiones vistas, podemos determinar:

\begin{equation}
\omega_0 = \frac{1}{CR_3}
\end{equation}

\begin{equation}
Q = \frac{R_2}{R_3}
\end{equation}

\begin{equation}
k = \frac{-R_1}{R_2}
\end{equation}

#### PUNTO #4:

Para normalizar la red vista, tomaremos a la resistencia $R_3$ como la norma de impedancia:

\begin{equation}
\Omega_Z = R_3
\end{equation}

\begin{equation}
R_2 = QR_3
\end{equation}

\begin{equation}
R_1 = R_3
\end{equation}

\begin{equation}
R_4 = R_3
\end{equation}

\begin{equation}
k = -1
\end{equation}

Por otro lado, la norma de frecuencia será $\omega_0$, por lo que:

\begin{equation}
\Omega_\omega = \omega_0 = 1 = \frac{1}{CR_3}
\end{equation}

\begin{equation}
C = \frac{1}{R_3}
\end{equation}

De esta manera, obtenemos la **Función Transferencia Normalizada**:

\begin{equation}
H(s) = \frac{k}{s^2 + s \frac{1}{Q} + 1}
\end{equation}

#### PUNTO #5:

Para la simulación en Python de la **Función Transferencia Normalizada**, el código utilizado y los resultados obtenidos fueron los siguientes:

```python
import matplotlib.pyplot as plt
import numpy as np
import scipy.signal as sig
from pytc2.sistemas_lineales import analyze_sys, pretty_print_lti

k = -1
wo = 1
qq = 1

num = np.array([k*(wo**2)]) 
den = np.array([1, wo/qq, wo**2])

pretty_print_lti(num,den)

H1 = sig.TransferFunction(num,den)

plt.show()
plt.close('all')
analyze_sys(H1, 'TS2')
```
