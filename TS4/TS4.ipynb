Para obtener las $T_2(s)$, $T_3(s)$ y $T_4(s)$, de órdenes 2, 3 y 4 respectivamente, se debe tener en cuenta el siguiente análisis matemático:

\begin{equation}
T(s) = e^{-s} = \frac{1}{e^s} = \frac{1}{senh(s) + cosh(s)} = \frac{\frac{1}{senh(s)}}{1 + cotgh(s)}
\end{equation}

Donde la $cotgh(s)$ se puede estimar por medio del método se Storch, mediante el cual se trunca su valor de acuerdo al orden del filtro buscado, a través de la siguiente expresión:

\begin{equation}
cotgh(s) = \frac{1}{s} + \frac{1}{\frac{3}{s} + \frac{1}{\frac{5}{s} + \frac{1}{\frac{7}{s} + ...}}}
\end{equation}

De esta manera, para los órdenes pedidos, las expresiones de cotgh(s) son las siguientes:

\begin{equation}
cotgh_2(s) = \frac{1}{s} + \frac{1}{\frac{3}{s}} = \frac{s^2 + 3}{3s}
\end{equation}

\begin{equation}
cotgh_3(s) = \frac{1}{s} + \frac{1}{\frac{3}{s} + \frac{1}{\frac{5}{s}}} = \frac{6s^2 + 15}{s^3 + 15s}
\end{equation}

\begin{equation}
cotgh_4(s) = \frac{1}{s} + \frac{1}{\frac{3}{s} + \frac{1}{\frac{5}{s} + \frac{1}{\frac{7}{s}}}} = \frac{s^4 + 45s^2 + 105}{10s^3 + 105s}
\end{equation}

Teniendo en cuenta que:

\begin{equation}
cotgh(s) = \frac{cosh(s)}{senh(s)}
\end{equation}

Ya podemos definir las Funciones Transferencia de Bessel de orden 2, 3 y 4:

\begin{equation}
T_2(s) = \frac{3}{s^2 + 3s + 3} = \frac{3}{(s + 1.5 - j0.866) \cdot (s + 1.5 + j0.866)}
\end{equation}

\begin{equation}
T_3(s) = \frac{15}{s^3 + 6s^2 + 15s + 15} = \frac{2.322}{s + 2.322} \cdot \frac{6.459}{(s + 1.839 - j1.754) \cdot (s + 1.839 + j1.754)}
\end{equation}

\begin{equation}
T_4(s) = \frac{105}{s^4 + 10s^3 + 45s^2 + 105s + 105} = \frac{11.487}{(s + 2.104 - j2.657) \cdot (s + 2.104 + j2.657)} \cdot \frac{9.139}{(s + 2.896 - j0.867) \cdot (s + 2.896 + j0.867)}
\end{equation}

#### PUNTO #2:

En este punto tendremos en cuenta la siguiente expresión:

\begin{equation}
T_n(s) = \frac{k \cdot \omega_0^2}{B_n(s)}
\end{equation}

Donde k = 1 y $B_n(s)$ es el polinomio de Bessel de orden "n". Luego, se plantea lo siguiente:

\begin{equation}
T(j\omega) = T(s)|_{s = j\omega} = \frac{\omega_0^2}{B_n(j\omega)}
\end{equation}

\begin{equation}
|T(j\omega)| = \frac{|\omega_0^2|}{|B_n(j\omega)|}
\end{equation}

\begin{equation}
\alpha(\omega) = \frac{1}{|T(j\omega)|} = \frac{|B_n(j\omega)|}{|\omega_0^2|}
\end{equation}

\begin{equation}
\alpha[dB] = 20log \left(\frac{1}{|T(j\omega)|} \right) = 20log \left(\frac{|B_n(j\omega)|}{|\omega_0^2|} \right)
\end{equation}

En esta última expresión, si $\omega = 1$ entonces el valor de la atenuación en dB coincide con el valor de atenuación máximo $\alpha_{MAX}$ pedido en el enunciado. De esta forma, buscaremos el módulo de cada una de las transferencias encontradas en el punto **#1** e igualaremos $\omega = 1$, calculando la atenuación, hasta que $\alpha_n \leq \alpha_{MAX}$

\begin{equation}
\rightarrow |T_2(j\omega)| = \frac{3}{|(j\omega)^2 + 3(j\omega) + 3|} = \frac{3}{\sqrt{(3 - \omega^2)^2 + (3\omega)^2}}
\end{equation}

\begin{equation}
\alpha_2[dB] = 20log \left(\frac{1}{|T_2(j\omega)|}|_{\omega = 1} \right) = 1.597 dB > \alpha_{MAX}
\end{equation}

\begin{equation}
\rightarrow |T_3(j\omega)| = \frac{15}{|(j\omega)^3 + 6(j\omega)^2 + 15(j\omega) + 15|} = \frac{15}{\sqrt{(15 - 6\omega^2)^2 + (15\omega - \omega^3)^2}}
\end{equation}

\begin{equation}
\alpha_3[dB] = 20log \left(\frac{1}{|T_3(j\omega)|}|_{\omega = 1} \right) = 0.903 dB < \alpha_{MAX}
\end{equation}

\begin{equation}
\rightarrow |T_4(j\omega)| = \frac{105}{|(j\omega)^4 + 10(j\omega)^3 + 45(j\omega)^2 + 105(j\omega) + 105|} = \frac{105}{\sqrt{(105 + \omega^4 - 45\omega^2)^2 + (105\omega - 10\omega^3)^2}}
\end{equation}

\begin{equation}
\alpha_4[dB] = 20log \left(\frac{1}{|T_4(j\omega)|}|_{\omega = 1} \right) = 0.62995 dB < \alpha_{MAX}
\end{equation}

Como se puede observar, el menor orden que nos permite cumplir con el valor de atenuación máxima pedido es n = 3.

#### PUNTO #3:

Ya conociendo la expresión de la transferencia de orden 3, es posible determinar las expresiones de su módulo y fase:

\begin{equation}
|T_3(j\omega)| = \frac{15}{\sqrt{(15 - 6\omega^2)^2 + (15\omega - \omega^3)^2}}
\end{equation}

\begin{equation}
\theta_3(\omega) = - tg^{-1} \left(\frac{15\omega - \omega^3}{15 - 6\omega^2} \right)

Por teoría, sabemos que la expresión del retardo de grupo se puede obtener derivando respecto de $\omega$ a la fase, y multiplicar el resultaro por -1. De esta manera, obtenemos:

\begin{equation}
D_3(\omega) = - \frac{d\theta_3(\omega)}{d\omega} = \frac{6\omega^4 + 45\omega^2 + 225}{\omega^6 + 6\omega^4 + 45\omega^2 + 225}
\end{equation}

\begin{equation}
D_3(\omega)|_{\omega = 2.5} = 0.752 [seg]
\end{equation}

Este valor del retardo de grupo, no da un porcentaje de desviación respecto de D(0) de %desv = 24.79%.

#### PUNTO #4:

A partir de la red planteada en el enunciado, se puede determinar la función transferencia de la misma para un k = 1:

\begin{equation}
T_{Sallen-Key}(s) = \frac{V_o}{V_i} = \frac{\frac{1}{C_1C_2R_1R_2}}{s^2 + s \frac{1}{C_1} \cdot (\frac{1}{R_1} + \frac{1}{R_2}) + \frac{1}{C_1C_2R_1R_2}}
\end{equation}

Agregando una red RC serie a la salida de la red tipo Sallen-Key vista, podemos obtener un filtro de orden 3 con la siguiente estructura y transferencia:

![alt text](red_propuesta_ts4.png)

\begin{equation}
T(s) = \frac{V_o}{V_i} = \frac{\frac{1}{C_1C_2R_1R_2}}{s^2 + s \frac{1}{C_1} \cdot (\frac{1}{R_1} + \frac{1}{R_2}) + \frac{1}{C_1C_2R_1R_2}} \cdot \frac{\frac{1}{C_3R_3}}{s + \frac{1}{C_3R_3}}
\end{equation}

Si comparamos esta última expresión, con la de la función transferencia de orden 3 encontrada en el punto **#1**, podemos determinar los siguientes datos:

\begin{equation}
T_3(s) = \frac{15}{s^3 + 6s^2 + 15s + 15} = \frac{2.322}{s + 2.322} \cdot \frac{6.459}{(s + 1.839 - j1.754) \cdot (s + 1.839 + j1.754)} = \frac{2.542^2}{s^2 + s \cdot \frac{2.542}{0.691} + 2.542^2} \cdot \frac{2.322}{s + 2.322}
\end{equation}

\begin{equation}
\rightarrow \omega_{01}^2 = 2.542^2 = \frac{1}{C_1C_2R_1R_2}
\end{equation}

\begin{equation}
\rightarrow \frac{\omega_{01}}{Q_1} = \frac{2.542}{0.691} = \frac{1}{C_1} \cdot (\frac{1}{R_1} + \frac{1}{R_2})
\end{equation}

\begin{equation}
\rightarrow \omega_{02} = 2.322 = \frac{1}{C_3R_3}
\end{equation}

Para poder llegar a los valores de los componentes, se decidió tomar como norma de impedancia a $\Omega_Z = R_1$, y como norma de frecuencia $\Omega_\omega = \omega_c$. De esta forma, los capacitores quedaron con los siguientes valores:

\begin{equation}
R_2 = R_3 = R_1
\end{equation}

\begin{equation}
\rightarrow C_1 = \frac{2Q_1}{\omega_{01} \cdot \Omega_Z \cdot \Omega_\omega}
\end{equation}

\begin{equation}
\rightarrow C_2 = \frac{1}{2Q_1 \cdot \omega_{01} \cdot \Omega_Z \cdot \Omega_\omega}
\end{equation}

\begin{equation}
\rightarrow C_3 = \frac{1}{\omega_{02} \cdot \Omega_Z \cdot \Omega_\omega}
\end{equation}

Como estamos buscando sintetizar el la red normalizada, tomamos $\Omega_Z = R_1 = 1$ y $\Omega_\omega = \omega_c = \frac{1}{D(0)} = 1$:

\begin{equation}
\rightarrow C_1 = \frac{2Q_1}{\omega_{01}}
\end{equation}

\begin{equation}
\rightarrow C_2 = \frac{1}{2Q_1 \cdot \omega_{01}}
\end{equation}

\begin{equation}
\rightarrow C_3 = \frac{1}{\omega_{02}}
\end{equation}
