La matriz obtenida representa al cuadripolo total, incluyendo el nodo "4". Para obtener la matriz pedida, será necesario considerar como interno dicho nodo, por medio del método de *Supresión de Polos*:

\begin{equation}
i' = Y_{MAI}' \cdot v'
\end{equation}

\begin{equation}
\begin{pmatrix}
I_1 \\
I_2 \\
I_3 \\
0
\end{pmatrix}       =       \begin{pmatrix}
                            Y_{11} & Y_{12} \\
                            Y_{21} & Y_{22}

\end{equation}

\begin{equation}
\begin{pmatrix}
i' \\
0
\end{pmatrix}       =       \begin{pmatrix}
                            Y_{11} & Y_{12} \\
                            Y_{21} & Y_{22}
                            \end{pmatrix}                                                   \cdot       \begin{pmatrix}
                                                                                                        v' \\
                                                                                                        v_d
                                                                                                        \end{pmatrix}
\end{equation}

Esto último nos deja con dos expresiones, de las surge la $Y_{MAI}'$:

\begin{equation}
i' = Y_{11} \cdot v' + Y_{12} \cdot v_d
\end{equation}

\begin{equation}
0 = Y_{21} \cdot v' + Y_{22} \cdot v_d
\end{equation}

\begin{equation}
\rightarrow i' = (Y_{11} - Y_{12} \cdot Y_{22}^{-1} \cdot Y_{21}) \cdot v'
\end{equation}

\begin{equation}
\rightarrow Y_{MAI}' = Y_{11} - Y_{12} \cdot Y_{22}^{-1} \cdot Y_{21}
\end{equation}

De esta forma, realizando la partición de la $Y_{MAI}$ original, se obtiene:

\begin{equation}
Y_{11} =   \begin{pmatrix}
            \frac{1}{sL} & 0 & 0 \\
            0 & \frac{1}{sL} & 0 \\
            0 & 0 & sC
            \end{pmatrix}
\end{equation}

\begin{equation}
Y_{12} =   \begin{pmatrix}
            \frac{-1}{sL} \\
            \frac{-1}{sL} \\
            -sC
            \end{pmatrix}
\end{equation}

\begin{equation}
Y_{21} =   \begin{pmatrix}
            \frac{-1}{sL} & \frac{-1}{sL} & -sC
            \end{pmatrix}
\end{equation}

\begin{equation}
Y_{22} =   \begin{pmatrix}
            \frac{2}{sL} + sC
            \end{pmatrix}       \rightarrow     Y_{22}^{-1} = \begin{pmatrix}
                                                                \frac{1}{\frac{2}{sL} + sC}
                                                                \end{pmatrix}
\end{equation}

Operando matricialmente, finalmente se obtiene:

\begin{equation}
Y_{MAI}' = \begin{pmatrix}
            \frac{1}{sL} (1-\frac{1}{s^2LC+2}) & -\frac{\frac{1}{sL}}{s^2LC+2} & -\frac{sC}{s^2LC+2} \\
            -\frac{\frac{1}{sL}}{s^2LC+2} & \frac{1}{sL} (1-\frac{1}{s^2LC+2}) & -\frac{sC}{s^2LC+2} \\
            -\frac{sC}{s^2LC+2} & -\frac{sC}{s^2LC+2} & sC (1-\frac{s^2LC}{s^2LC+2})
            \end{pmatrix}
\end{equation}

#### PUNTO #2:

**(a)** Para este inciso, utilizaremos las siguientes expresiones, representativas de una trasferencia de tensiones y de cofactores de segundo orden, respectivamente:

\begin{equation}
\rightarrow T(s) = \frac{V_{ij}}{V_{mn}} = (-1)^{i+j+m+n} \cdot \frac{Y/_{ij}^{mn}}{Y/_{mn}^{mn}}
\end{equation}

\begin{equation}
\rightarrow Y/_{ij}^{mn} = (-1)^{i+j+m+n} \cdot |Y_{ij}^{mn}|
\end{equation}

De esta forma, a continuación se detalla la obtención de cada cofactor de segundo orden, y la transferencia resultante:

\begin{equation}
Y/_{23}^{13} = (-1)^{2+3+1+3} \cdot |Y_{23}^{13}| = \frac{\frac{1}{sL}}{s^2LC+2}
\end{equation}

\begin{equation}
Y/_{13}^{13} = (-1)^{1+3+1+3} \cdot |Y_{13}^{13}| = \frac{1}{sL} \cdot \frac{s^2LC+1}{s^2LC+2}
\end{equation}

\begin{equation}
\rightarrow T(s)_a = \frac{V_{23}}{V_{13}} = (-1)^{2+3+1+3} \cdot \frac{Y/_{23}^{13}}{Y/_{13}^{13}} = (-1) \cdot \frac{\frac{1}{LC}}{s^2 + \frac{1}{LC}}
\end{equation}

----

\begin{equation}
Y/_{21}^{31} = (-1)^{2+1+3+1} \cdot |Y_{21}^{31}| = \frac{sC}{s^2LC+2}
\end{equation}

\begin{equation}
Y/_{31}^{31} = (-1)^{3+1+3+1} \cdot |Y_{31}^{31}| = \frac{1}{sL} \cdot \frac{s^2LC+1}{s^2LC+2}
\end{equation}

\begin{equation}
\rightarrow T(s)_b = \frac{V_{21}}{V_{31}} = (-1)^{2+1+3+1} \cdot \frac{Y/_{21}^{31}}{Y/_{31}^{31}} = (-1) \cdot \frac{s^2}{s^2 + \frac{1}{LC}}
\end{equation}

----

\begin{equation}
Y/_{12}^{32} = (-1)^{1+2+3+2} \cdot |Y_{12}^{32}| = \frac{-sC}{s^2LC+2}
\end{equation}

\begin{equation}
Y/_{32}^{32} = (-1)^{3+2+3+2} \cdot |Y_{32}^{32}| = \frac{1}{sL} \cdot \frac{s^2LC+1}{s^2LC+2}
\end{equation}

\begin{equation}
\rightarrow T(s)_c = \frac{V_{12}}{V_{32}} = (-1)^{1+2+3+2} \cdot \frac{Y/_{12}^{32}}{Y/_{32}^{32}} = (-1) \cdot \frac{s^2}{s^2 + \frac{1}{LC}}
\end{equation}

**(b)** Por último, para calcular los parámetros T o ABCD, realizamos una transformación de los parámetros $Y_S$ encontrados en el inciso anterior:

-----

$\rightarrow$ **PARÁMETROS Y A PARÁMETROS T**

\begin{equation}
Y = \begin{pmatrix}
    Y_{11} & Y_{12} \\
    Y_{21} & Y_{22}
    \end{pmatrix}   \rightarrow     T = \begin{pmatrix}
                                        A & B \\
                                        C & D
                                        \end{pmatrix}   =  \frac{-1}{Y_{21}} \begin{pmatrix}
                                                                            Y_{22} & 1 \\
                                                                            \Delta Y & Y_{11}
                                                                            \end{pmatrix}
\end{equation}

----

\begin{equation}
Y_{11} = sC_{iss} \quad ; \quad Y_{12} = -sC_{rss} \quad ; \quad Y_{21} = gm - sC_{rss} \quad ; \quad Y_{22} = sC_{oss}
\end{equation}

\begin{equation}
\Delta Y_S = Y_{11} cdot Y_{22} - Y_{21} cdot Y_{12} = s^2 \cdot (C_{rss}^2 + C_{iss} \cdot C_{oss}) - s \cdot gmC_{rss}
\end{equation}

\begin{equation}
T = \begin{pmatrix}
    A & B \\
    C & D
    \end{pmatrix}   =  \frac{1}{sC_{rss} - gm} \begin{pmatrix}
                                                sC_{oss} & 1 \\
                                                s^2 \cdot (C_{rss}^2 + C_{iss} \cdot C_{oss}) - s \cdot gmC_{rss} & sC_{iss}
                                                \end{pmatrix}
\end{equation}
